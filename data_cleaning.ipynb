```python
import pandas as pd
```

```python
data = pd.read_csv('earphones.csv')
data.head()
```

### Remove In the ear in the Color

```python
color = data['Color']
for i , value in enumerate(color) :
    value = value.split(',')[:-1]
    color[i] = ', '.join(value)
#data['Color'] = color 
data.head(10)
```

### Make Price as integer


```python
price = data['Price']
price.head()
for i , value in enumerate(price) :
    temp = value.replace(',' , '')
    price[i] = int(temp[1:])
    
#data['Price'] = price
data.head(10)
```

### Discount cleaning

```python
discount = data["Discount"]
for i , value in enumerate(discount) :
    discount[i] = str(value).replace("off" , '')
    
#data['Discount'] = discount
data.head(10)
```

### Rounded value of Buyers

```python
buyers = data["Buyers"]
for i , value in enumerate(buyers):
    buyers[i] = str(value)[1:-1]
#data['Buyers'] = buyers
data.head(10)
```

## Exporting the Cleaned file

```python
data.to_csv("cleaned_data.csv")
```
